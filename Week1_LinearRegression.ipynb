```python
#imports
import numpy as np
import matplotlib.pyplot as plt

#random seed 
rng = np.random.default_rng(21)
```

```python
# Generate test data

# size
n = 1000

# n x-values between 1 and 100
x = rng.integers(1, 100, n)
x.sort()

# define linear trend
slope = 2
intercept = 2
y_trend = slope * x + intercept

# add noise
noise = rng.normal(0, 20, n)
y = y_trend + noise

# plot
plt.scatter(x,y, label="Data with noise")
plt.plot(x, y_trend, color="red", label="True trend")
plt.legend()
plt.show()
```

```python
# Regressionsgleichung aufstellen

# mean and standard deviation
m_x = x.sum()/n
m_y = y.sum()/n

s_x = np.sqrt(np.square(x-m_x).sum()/(n-1))
s_y = np.sqrt(np.square(y-m_y).sum()/(n-1))

# covariance and correlation
kov = ((x-m_x)*(y-m_y)).sum()/(n-1)
kor = (kov)/(s_x*s_y)

b = (s_y / s_x)*kor
a = -(s_y/s_x)*kor*m_x+m_y
reg_x = np.linspace(1,100)
reg_y = b * reg_x + a

plt.plot(reg_x, reg_y, color="green", label="regression")
plt.scatter(x,y, label="Data with noise")
plt.plot(x, y_trend, color="red", label="True trend")
plt.legend()
plt.show()
print(m_x, s_x, m_y, s_y, kov, kor, a, b)
```
